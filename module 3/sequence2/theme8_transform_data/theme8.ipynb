### How to engineer features in the data.

#### New variables

```python
#New Variable
from dateutil.relativedelta import *
from datetime import *
  
def get_age(dob):
    now = datetime.now()
    age = relativedelta(now, dob).years
    return age
  
df['age'] = pd.to_datetime(df['dob']).apply(get_age)
  
df
```

#### Replace values

```python
obfuscated = df.copy()
obfuscated['name'] = obfuscated['name'].replace(to_replace = r'\s(.*)', value = ' LASTNAME', regex=True)
obfuscated
```

#### Splitting and combining

```python
splitnames = df.copy()
split = splitnames['name'].str.split(' ', expand = True)
splitnames['first'] = split[0]
splitnames['last'] = split[1]
splitnames
```

```python
#Combining 
splitnames['lastfirst'] = splitnames['last'] + ', ' + splitnames['first']
splitnames 
```

### How to transform data values.

```python
df['target'].hist()
```

#### Distribution transforms

```python
#Log transform
transforms = df.copy()
transforms['log'] = transforms['target'].transform(np.log10)
transforms['log'].hist()
```

```python
#Cube transform
transforms = df.copy()
transforms['cube'] = transforms['target'].transform(lambda x: np.power(x, 3))
transforms['cube'].hist()
```

#### Scaling transforms.

```python
#Normalizing
scaling = df.copy()
min_target = np.min(scaling['target'])
max_target = np.max(scaling['target'])
scaling['norm_target'] = (scaling['target'] - min_target) / (max_target - min_target)
scaling
```

```python
#Standardizing
scaling = df.copy()
mean_target = np.mean(scaling['target'])
sd_target = np.std(scaling['target'])
scaling['standardized_target'] = (scaling['target'] - mean_target) / (sd_target)
  
scaling
```

# Combining Data in pandas With merge(), .join(), and concat()

### 1. pandas merge() : Combiner des données sur des colonnes ou indices communs

merge() est utile lorsque vous voulez une fonctionnalité similaire aux opérations de jointure d'une base de données. C'est l'opération la plus flexible des trois.

```python
# Création de deux DataFrames simples
df1 = pd.DataFrame({
    'id': [1, 2, 3],
    'nom': ['Alice', 'Bob', 'Charlie'],
    'age': [25, 30, 35]
})

df2 = pd.DataFrame({
    'id': [2, 3, 4],
    'ville': ['Paris', 'Lyon', 'Marseille'],
    'salaire': [50000, 60000, 45000]
})

# Fusion sur la colonne 'id'
resultat = pd.merge(df1, df2, on='id')
print(resultat)
```

#### Types de jointures

1 - Jointure interne (inner) : Ne conserve que les lignes avec des clés présentes dans les deux DataFrames

```python
resultat = pd.merge(df1, df2, on='id', how='inner')
resultat
```

2 - Jointure externe (outer) : Conserve toutes les lignes des deux DataFrames

```python
r = pd.merge(df1, df2, on='id', how='outer')
r
```

3 -     Jointure gauche (left) : Conserve toutes les lignes du DataFrame de gauche

```python
r = pd.merge(df1, df2, on='id', how='left')
r
```

4 -     Jointure droite (right) : Conserve toutes les lignes du DataFrame de droite

```python
r = pd.merge(df1, df2, on='id', how='right')
r
```

```python
employes = pd.DataFrame({
    'id_employe': [101, 102, 103, 104],
    'nom': ['Dupont', 'Martin', 'Durand', 'Leroy'],
    'id_dept': [1, 2, 1, 3]
})

departements = pd.DataFrame({
    'id_dept': [1, 2, 4],
    'nom_dept': ['Ventes', 'Marketing', 'IT'],
    'localisation': ['Paris', 'Lyon', 'Toulouse']
})

# Jointure gauche pour conserver tous les employés
jointure = pd.merge(employes, departements, on='id_dept', how='left')
print(jointure)
```

### 2. pandas join() : Combiner des données sur une colonne ou un index

join() est une méthode d'instance qui simplifie certaines opérations de fusion.

```python
# Reprenons les DataFrames précédents
# On doit d'abord définir l'index
employes_avec_index = employes.set_index('id_dept')
departements_avec_index = departements.set_index('id_dept')

# Jointure simple
resultat = employes_avec_index.join(departements_avec_index, lsuffix='_emp', rsuffix='_dept')
print(resultat)
```

### 3. pandas concat() : Combiner des données en lignes ou colonnes

```python
# Création de deux DataFrames similaires
df_a = pd.DataFrame({
    'A': ['A0', 'A1'],
    'B': ['B0', 'B1']
})

df_b = pd.DataFrame({
    'A': ['A2', 'A3'],
    'B': ['B2', 'B3']
})

# Concaténation verticale (par défaut)
resultat = pd.concat([df_a, df_b])
print(resultat)
```

```python
df_c = pd.DataFrame({
    'C': ['C0', 'C1'],
    'D': ['D0', 'D1']
})

resultat = pd.concat([df_a, df_c], axis=1)
print(resultat)
```
